```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams["font.size"] = 12
```

```python
# ===================================
# Radar Parameters
# ===================================

fs = 20e6
T = 20e-6
B = 5e6

N = int(fs*T)

k = B/T

print("N =", N)
```

```python
t = np.arange(N)/fs

t = t - T/2

s = np.exp(1j*np.pi*k*t**2)
```

```python
def delay_signal(signal, delay):

    delayed = np.zeros_like(signal)

    delayed[delay:] = signal[:-delay]

    return delayed
```

```python
taus = [30,100,220]

signals = []

for tau in taus:

    signals.append(
        delay_signal(s,tau)
    )

S = np.column_stack(signals)

print(S.shape)
```
