## Delay Model

Goal

Generate delayed waveform

Construct signal matrix S

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def delay_signal(signal, delay):

    delayed = np.zeros_like(signal)

    delayed[delay:] = signal[:-delay]

    return delayed
```

```python
fs = 20e6              # Sampling frequency (Hz)

T = 20e-6              # Pulse duration

B = 5e6                # Chirp bandwidth

f0 = -B/2              # Start frequency

f1 = B/2               # End frequency

k = B / T

N = int(T * fs)

t = np.arange(N) / fs

t = t - T/2

s = np.exp(1j * np.pi * k * t**2)
```

```python
tau = 50

s2 = delay_signal(s, tau)
```

```python
plt.figure(figsize=(10,3))

plt.plot(t * 1e6, np.real(s2), label="Real")
plt.plot(t * 1e6, np.imag(s2), label="Imag")
plt.legend()

plt.xlabel("Time (us)")
plt.ylabel("Amplitude")

plt.title("Delayed LFM Waveform")

plt.grid(True)

plt.show()
```
